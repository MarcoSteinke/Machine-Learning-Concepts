# Numpy basics:


```python
import numpy as np
import random as random

# arrange 100 nums
np.arange(100)

# reshape 100 nums into 50 tuples of size 2
np.arange(100).reshape(50,2)

data = np.arange(16).reshape(4,4)

# print shape of the data
data.shape

# print dimensions of the data
data.ndim

# find data type
data.dtype

# find size a an item inside the data
data.itemsize

# transform array to np array
more_data = np.array([1,2,3])
type(more_data)
```

## Numpy arrays:

```python
list = [(1, 2), (3, 4), (5, 6), (7, 8)]
list_as_array = np.array(list)
list_as_array

# transform into complex array
list_as_complex_array = np.array(list, dtype=complex)
list_as_complex_array

# get array of zeros
np.zeros((3,4))

np.ones((10,10))

# generate numbers with a certain distance between
np.arange(10, 21, 5)

# get a series of numbers (9 values between 0 and 2)
np.linspace(0,2, 9)

from numpy import pi

# get 100 numbers between 0 and 2 * pi
series = np.linspace(-2 * pi, 2 * pi, 100)

import matplotlib.pyplot as plt
import math

cos_values = [math.cos(x) for x in series.tolist()]

plt.plot(cos_values)
```

Numpy does offer its own methods to very easily map the values inside of a np.array to other values.

```python
# the following method is much simpler
series = np.linspace(-2 * pi, 2 * pi, 100)
cos_values = np.cos(series)

plt.plot(cos_values)
```
